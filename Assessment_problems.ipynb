```python
class Passenger:
    def __init__(self, name):
        self.name = name
class Flight:
    def __init__(self, flight_number):
        self.flight_number = flight_number
        self.occupied_seats = set()
    def assign_seat(self, seat_number):
        if seat_number in self.occupied_seats:
            return False
        self.occupied_seats.add(seat_number)
        return True
class BoardingPass:
    def __init__(self, passenger, flight, seat_number):
        self.passenger = passenger
        self.flight = flight
        self.seat_number = seat_number
    def generate_boarding_pass(self):
        print("==================================================")
        print("               BOARDING PASS")
        print("==================================================")
        print(f"Passenger Name : {self.passenger.name}")
        print(f"Flight Number  : {self.flight.flight_number}")
        print(f"Seat Number    : {self.seat_number}")
        print("Status         : CHECK-IN COMPLETE")
        print("==================================================")

class Airport:
    def __init__(self):
        self.passengers = []

    def register_passenger(self, passenger_name):
        passenger = Passenger(passenger_name)
        self.passengers.append(passenger)
        return passenger

airport = Airport()
passenger = airport.register_passenger("Omkar")
flight = Flight(1587)
seat = "12A"
if flight.assign_seat(seat):
    boarding_pass = BoardingPass(passenger, flight, seat)
    boarding_pass.generate_boarding_pass()
```

```python
class Vehicle:
    def __init__(self, vehicle_number):
        self.vehicle_number = vehicle_number

class ParkingSlot:
    def __init__(self, slot_id, rate_per_hour):
        self.slot_id = slot_id
        self.rate_per_hour = rate_per_hour

class ParkingManager:
    def __init__(self, vehicle, parking_slot, hours_parked):
        self.vehicle = vehicle
        self.parking_slot = parking_slot
        self.hours_parked = hours_parked

    def calculate_fee(self):
        return self.hours_parked * self.parking_slot.rate_per_hour

    def generate_receipt(self):
        fee = self.calculate_fee()
        print("==================================================")
        print("              PARKING RECEIPT")
        print("==================================================")
        print(f"Vehicle Number : {self.vehicle.vehicle_number}")
        print(f"Hours Parked   : {self.hours_parked}")
        print(f"Parking Fee    : {fee}/-")
        print("==================================================")

car = Vehicle("AP39AB1234")
slot = ParkingSlot(slot_id="A1", rate_per_hour=30)
manager = ParkingManager(car, slot, hours_parked=5)
manager.generate_receipt()
```

```python
class Player:
    def __init__(self, name):
        self.name = name
        self.matches = []

    def add_match_performance(self, match):
        self.matches.append(match)

    def calculate_total_runs(self):
        return sum(match.runs for match in self.matches)

    def calculate_average_runs(self):
        total_matches = len(self.matches)
        if total_matches == 0:
            return 0.0
        return self.calculate_total_runs() / total_matches

class Match:
    def __init__(self, match_id, runs):
        self.match_id = match_id
        self.runs = runs

class Tournament:
    def __init__(self, name):
        self.name = name

    def generate_performance_report(self, player):
        print("==================================================")
        print("             PLAYER PERFORMANCE REPORT")
        print("==================================================")
        print(f"Player Name    : {player.name}")
        print(f"Total Runs     : {player.calculate_total_runs()}")
        print(f"Matches Played : {len(player.matches)}")
        print(f"Average Runs   : {player.calculate_average_runs():.1f}")
        print("==================================================")

player_arjun = Player("Omkar")

player_arjun.add_match_performance(Match("Match 1", 50))
player_arjun.add_match_performance(Match("Match 2", 75))
player_arjun.add_match_performance(Match("Match 3", 100))

tournament = Tournament("Academy Championship")
tournament.generate_performance_report(player_arjun)
```

```python
class Guest:
    def __init__(self, name):
        self.name = name

class Room:
    def __init__(self, room_type, room_rate):
        self.room_type = room_type
        self.room_rate = room_rate

class Reservation:
    def __init__(self, guest, room, nights):
        self.guest = guest
        self.room = room
        self.nights = nights

    def calculate_total_amount(self):
        return self.room.room_rate * self.nights

    def generate_invoice(self):
        total = self.calculate_total_amount()
        print("==================================================")
        print("              HOTEL INVOICE")
        print("==================================================")
        print(f"Guest Name     : {self.guest.name}")
        print(f"Room Type      : {self.room.room_type}")
        print(f"Nights Stayed  : {self.nights}")
        print(f"Total Amount   : ₹{total}")
        print("==================================================")

guest_sophia = Guest("Omkar")
deluxe_room = Room("Deluxe", 3500)
booking = Reservation(guest_sophia, deluxe_room, nights=5)

booking.generate_invoice()
```

```python
class Customer:
    def __init__(self, name, mobile_number):
        self.name = name
        self.mobile_number = mobile_number

class RechargePlan:
    def __init__(self, plan_name, amount):
        self.plan_name = plan_name
        self.amount = amount

class Recharge:
    def __init__(self, customer):
        self.customer = customer
        self.selected_plan = None

    def select_plan(self, plan):
        self.selected_plan = plan

    def generate_receipt(self):
        if self.selected_plan is None:
            print("No plan selected.")
            return

        print("==================================================")
        print("             RECHARGE RECEIPT")
        print("==================================================")
        print(f"Customer Name : {self.customer.name}")
        print(f"Plan Selected : {self.selected_plan.plan_name}")
        print(f"Amount Paid   : ₹{self.selected_plan.amount}")
        print("Status        : SUCCESSFUL")
        print("==================================================")

customer_logan = Customer("Omkar", "6302207891")
plan_unlimited = RechargePlan("Unlimited 84 Days", 899)
recharge_transaction = Recharge(customer_logan)
recharge_transaction.select_plan(plan_unlimited)
recharge_transaction.generate_receipt()
```

```python
class Client:
    def __init__(self, name):
        self.name = name

class Project:
    def __init__(self, project_name, hours_worked, hourly_rate):
        self.project_name = project_name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

class Invoice:
    def __init__(self, client, project):
        self.client = client
        self.project = project

    def calculate_invoice_amount(self):
        return self.project.hours_worked * self.project.hourly_rate

    def generate_invoice(self):
        amount = self.calculate_invoice_amount()
        print("==================================================")
        print("                CLIENT INVOICE")
        print("==================================================")
        print(f"Client Name   : {self.client.name}")
        print(f"Project Name  : {self.project.project_name}")
        print(f"Hours Worked  : {self.project.hours_worked}")
        print(f"Hourly Rate   : ₹{self.project.hourly_rate}")
        print(f"Invoice Amount: ₹{amount}")
        print("==================================================")

client_olivia = Client("Olivia")
portfolio_project = Project("Portfolio Website", hours_worked=20, hourly_rate=1000)
client_invoice = Invoice(client_olivia, portfolio_project)

client_invoice.generate_invoice()
```

```python
class Pet:
    def __init__(self, name):
        self.name = name
class Owner:
    def __init__(self, name):
        self.name = name
class Appointment:
    def __init__(self, owner, pet, service, charge):
        self.owner = owner
        self.pet = pet
        self.service = service
        self.charge = charge

    def generate_receipt(self):
        print("==================================================")
        print("            SERVICE RECEIPT")
        print("==================================================")
        print(f"Owner Name : {self.owner.name}")
        print(f"Pet Name   : {self.pet.name}")
        print(f"Service    : {self.service}")
        print(f"Amount     : ₹{self.charge}")
        print("==================================================")

class PetCareCenter:
    def __init__(self):
        self.appointments = []
    def schedule_appointment(self, appointment):
        self.appointments.append(appointment)
        return appointment
owner_ethan = Owner("Ethan")
pet_bruno = Pet("Bruno")
center = PetCareCenter()
appt = Appointment(owner_ethan, pet_bruno, service="Grooming", charge=800)
center.schedule_appointment(appt)
appt.generate_receipt()
```

```python
class Match:
    def __init__(self, match_id, score):
        self.match_id = match_id
        self.score = score

class Player:
    def __init__(self, name):
        self.name = name
        self.matches = []

    def add_match_score(self, match):
        self.matches.append(match)

    def calculate_final_score(self):
        return sum(match.score for match in self.matches)

    def get_rank_status(self):
        if self.calculate_final_score() >= 300:
            return "QUALIFIED"
        return "ELIMINATED"

class Tournament:
    def __init__(self, name):
        self.name = name

    def generate_report(self, player):
        print("==================================================")
        print("            TOURNAMENT REPORT")
        print("==================================================")
        print(f"Player Name : {player.name}")
        print(f"Final Score : {player.calculate_final_score()}")
        print(f"Rank Status : {player.get_rank_status()}")
        print("==================================================")
player_leo = Player("Leo")
player_leo.add_match_score(Match("Match 1", 100))
player_leo.add_match_score(Match("Match 2", 150))
player_leo.add_match_score(Match("Match 3", 200))

esports_tourney = Tournament("Apex Championship")
esports_tourney.generate_report(player_leo)
```

```python
class Member:
    def __init__(self, name):
        self.name = name

class MembershipPlan:
    def __init__(self, plan_name, monthly_fee):
        self.plan_name = plan_name
        self.monthly_fee = monthly_fee

class Gym:
    def __init__(self, member, plan, duration_months):
        self.member = member
        self.plan = plan
        self.duration_months = duration_months

    def calculate_total_fee(self):
        return self.plan.monthly_fee * self.duration_months

    def generate_receipt(self):
        total_fee = self.calculate_total_fee()
        print("==================================================")
        print("             MEMBERSHIP RECEIPT")
        print("==================================================")
        print(f"Member Name : {self.member.name}")
        print(f"Plan        : {self.plan.plan_name}")
        print(f"Duration    : {self.duration_months} Months")
        print(f"Total Fee   : ₹{total_fee}")
        print("==================================================")
member_ava = Member("Ava")
premium_plan = MembershipPlan("Premium", monthly_fee=2000)
gym_membership = Gym(member_ava, premium_plan, duration_months=6)
gym_membership.generate_receipt()
```

```python
class Transaction:
    def __init__(self, txn_type, amount, status="SUCCESSFUL"):
        self.txn_type = txn_type
        self.amount = amount
        self.status = status

class Wallet:
    def __init__(self, initial_balance=0.0):
        self.balance = initial_balance
        self.transactions = []
    def add_money(self, amount):
        if amount > 0:
            self.balance += amount
            self.transactions.append(Transaction("DEPOSIT", amount))
            return True
        return False

    def transfer_money(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            self.transactions.append(Transaction("TRANSFER", amount))
            return True
        self.transactions.append(Transaction("TRANSFER", amount, status="FAILED"))
        return False

    def show_balance(self):
        return self.balance

class User:
    def __init__(self, name, initial_balance):
        self.name = name
        self.wallet = Wallet(initial_balance)

    def process_transfer(self, amount):
        opening_balance = self.wallet.show_balance()
        if self.wallet.transfer_money(amount):
            current_balance = self.wallet.show_balance()
            self._print_receipt(opening_balance, amount, current_balance, "SUCCESSFUL")
        else:
            current_balance = self.wallet.show_balance()
            self._print_receipt(opening_balance, amount, current_balance, "FAILED")

    def _print_receipt(self, opening, transfer, current, status):
        print("==================================================")
        print("          TRANSACTION RECEIPT")
        print("==================================================")
        print(f"User Name       : {self.name}")
        print(f"Opening Balance : ₹{opening}")
        print(f"Transfer Amount : ₹{transfer}")
        print(f"Current Balance : ₹{current}")
        print(f"Status          : {status}")
        print("==================================================")
user_noah = User("Noah", initial_balance=10000)
user_noah.process_transfer(2500)
```
