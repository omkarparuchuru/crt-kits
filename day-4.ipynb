```python
#Ticket Price

size=int(input("enter the size of list:"))
age=[]
for i in range(size):
  ele=int(input("enter the age:"))
  age.append(ele)
print(age)
for i in age:
  if(i>=1 and i<=100):
    if(i<12):
      print(f"{i}--------------> $10")
    elif(i>=12 and i<=60):
      print(f"{i}--------------> $15")
    else:
      print(f"{i}--------------> $12")
```

```python
#Atm Transaction
acc_balance=0
pin=int(input("enter the pin:"))
if pin==1234:
  print("Welcome To The Bank")
while True:
  print("1. Deposit")
  print("2. Cash Withdrawal")
  print("3. Balance")
  print("4. Exit")
  choice=int(input("Enter your choice: "))
  print("\n")
  if(choice==1):
    amount=int(input("Enter the amount to deposit: "))
    acc_balance=acc_balance+amount
    print(f"Dear customer your account xxxxxxxxx1234 is credited with {amount}")
  elif(choice==2):
    amount=int(input("Enter the amount to withdraw: "))
    if amount<=acc_balance:
      acc_balance=acc_balance-amount
      print(f"Dear customer your account xxxxxxxxx1234 is debited with {amount}")
    else:
      print("Insufficient balance.")
  elif(choice==3):
    print(f"Dear customer your account xxxxxxxxx1234 has current balance of {acc_balance}")
  elif(choice==4):
    print("Thank you for using our services. Goodbye!")
    break
  else:
    print("Invalid choice. Please try again.")
```
