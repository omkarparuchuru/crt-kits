```python
class Candidate:
    def __init__(self, name, skills):
        self.name = name
        self.skills = set(skills)  # Set -> Candidate Skills


class JobOpening:
    def __init__(self, title, required_skills, min_score):
        self.title = title
        self.required_skills = set(required_skills)
        self.min_score = min_score


class Interview:
    def __init__(self, candidate, job):
        self.candidate = candidate
        self.job = job
        self.scores = {}  # Dictionary -> Interview Scores

    def add_score(self, round_name, score):
        self.scores[round_name] = score

    def is_selected(self):
        # Calculate average score
        avg_score = sum(self.scores.values()) / len(self.scores) if self.scores else 0
        # Check if candidate has required skills
        has_skills = self.job.required_skills.issubset(self.candidate.skills)

        return avg_score >= self.job.min_score and has_skills


# --- Testing the System ---
if __name__ == "__main__":
    # Create Job
    job = JobOpening("Python Developer", ["Python", "SQL"], min_score=75)

    # Create Candidate (List converted to Set)
    candidate1 = Candidate("Alice", ["Python", "SQL", "Git"])

    # Run Interview
    interview = Interview(candidate1, job)
    interview.add_score("Technical", 80)
    interview.add_score("HR", 85)

    # Check Result
    if interview.is_selected():
        print(f"🎉 {candidate1.name} is hired for {job.title}!")
    else:
        print(f"❌ {candidate1.name} did not meet the criteria.")
```

```python
class Candidate:
    def __init__(self, candidate_id: str, name: str, skills: list):
        self.candidate_id = candidate_id
        self.name = name
        self.skills = set(skills)
        self.applications = []

    def apply_for_job(self, job_id: str):
        if job_id not in self.applications:
            self.applications.append(job_id)


class JobOpening:
    def __init__(self, job_id: str, title: str, required_skills: list, min_score: float):
        self.job_id = job_id
        self.title = title
        self.required_skills = set(required_skills)
        self.min_score = min_score


class Interview:
    def __init__(self, interview_id: str, candidate_id: str, job_id: str):
        self.interview_id = interview_id
        self.candidate_id = candidate_id
        self.job_id = job_id
        self.scores = {}

    def record_score(self, round_name: str, score: float):
        self.scores[round_name] = score

    def calculate_overall_score(self) -> float:
        if not self.scores:
            return 0.0
        return sum(self.scores.values()) / len(self.scores)


class RecruitmentManager:
    def __init__(self):
        self.candidates = {}
        self.jobs = {}
        self.interviews = []

    def add_candidate(self, candidate: Candidate):
        self.candidates[candidate.candidate_id] = candidate

    def add_job(self, job: JobOpening):
        self.jobs[job.job_id] = job

    def schedule_interview(self, candidate_id: str, job_id: str) -> Interview:
        interview_id = f"INT-{len(self.interviews) + 1}"
        interview = Interview(interview_id, candidate_id, job_id)
        self.interviews.append(interview)
        return interview

    def generate_selection_report(self, job_id: str):
        job = self.jobs.get(job_id)
        if not job:
            print("Job opening not found.")
            return

        print(f"\n--- FINAL SELECTION REPORT FOR: {job.title} ({job_id}) ---")
        print(f"Required Skills: {', '.join(job.required_skills)}")
        print(f"Minimum Passing Score: {job.min_score}\n")

        has_selections = False

        for interview in self.interviews:
            if interview.job_id != job_id:
                continue

            candidate = self.candidates.get(interview.candidate_id)
            if not candidate:
                continue


            matching_skills = candidate.skills.intersection(job.required_skills)
            overall_score = interview.calculate_overall_score()


            if matching_skills and overall_score >= job.min_score:
                has_selections = True
                print(f" SELECTED: {candidate.name} (ID: {candidate.candidate_id})")
                print(f" Match Scores: {interview.scores}")
                print(f" Overall Score: {overall_score:.2f}")
                print(f" Skills Matched: {', '.join(matching_skills)}\n")

        if not has_selections:
            print("No candidates met the selection criteria for this role.")



if __name__ == "__main__":

    manager = RecruitmentManager()


    python_dev = JobOpening("JOB-01", "Python Developer", ["Python", "SQL", "Git"], min_score=75.0)
    manager.add_job(python_dev)


    alice = Candidate("CAND-01", "Alice Smith", ["Python", "Django", "SQL", "Git"])
    bob = Candidate("CAND-02", "Bob Jones", ["Java", "SQL"])

    manager.add_candidate(alice)
    manager.add_candidate(bob)

    alice.apply_for_job("JOB-01")
    bob.apply_for_job("JOB-01")


    alice_interview = manager.schedule_interview("CAND-01", "JOB-01")
    alice_interview.record_score("Technical", 85.0)
    alice_interview.record_score("HR", 80.0)

    bob_interview = manager.schedule_interview("CAND-02", "JOB-01")
    bob_interview.record_score("Technical", 60.0)
    bob_interview.record_score("HR", 70.0)


    manager.generate_selection_report("JOB-01")
```

```python
class Student:
  def __init__(self,sid,sname):
    self.sid=sid
    self.sname=sname
class Assessment:
  def __init__(self,att,ass_score):
    self.att=att
    self.ass_score=ass_score
class PlacementManager:
  def __init__(self):
    self.student_list=[]

  def add_student(self,student,assessment):
    record={
        "Students info":student,
        "Assessment":assessment
    }
    self.student_list.append(record)

  def check_eligibility(self,attendance,score):
    if attendance>=75 and score>=60:
      return "Eligible"
    else:
      return "Not Eligible"

  def generate_report(self):
    for record in self.student_list:
      student=record["Students info"]
      assessment=record["Assessment"]

      status=self.check_eligibility(assessment.att,assessment.ass_score)

      print("="*50)
      print("          PLACEMENT ELIGIBILITY REPORT")
      print("="*50)
      print(f"Student ID:         {student.sid}")
      print(f"Student Name:       {student.sname}")
      print(f"Attendance:         {assessment.att}")
      print(f"Assessment Score:   {assessment.ass_score}%")
      print(f"Placement Status:   {status}")
      print("="*50)
pm=PlacementManager()
s=Student(1,"Omkar")
ass=Assessment(85,78)
pm.add_student(s,ass)
pm.generate_report()
```
